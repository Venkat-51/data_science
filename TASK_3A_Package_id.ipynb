```python
grouped = data.groupby('Package ID')
```

```python
for package_id, group_data in grouped:
    print("A -", package_id)
    for index, row in group_data.iterrows():  # Fixed the typo "iterr" to "iterrows"
        print("B -", row['Location'])
        print("C -", row['Hotel ID'])
        print("D -", row['SN-01'])
        print("E -", row['Property'])
        print("F -", row['Room Category'])
        print("G -", row['IDR to INR (CP) - 1 Night'])
        print()
```

```python
import pandas as pd

# Assuming 'grouped' is your grouped DataFrame
# If 'grouped' is not defined, you need to define it based on your data

# Create an empty DataFrame to store the results
result_df = pd.DataFrame()

for package_id, group_data in grouped:
    temp_df = pd.DataFrame({
        'A - Package ID': [package_id] * len(group_data),
        'B - Location': group_data['Location'],
        'C - Hotel ID': group_data['Hotel ID'],
        'D - SN-01': group_data['SN-01'],
        'E - Property': group_data['Property'],
        'F - Room Category': group_data['Room Category'],
        'G - IDR to INR (CP) - 1 Night': group_data['IDR to INR (CP) - 1 Night']
    })
    result_df = pd.concat([result_df, temp_df], ignore_index=True)

# Writing the result DataFrame to Excel
result_df.to_excel('output.xlsx', index=False)
```
